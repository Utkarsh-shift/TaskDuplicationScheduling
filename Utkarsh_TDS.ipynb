```python
def ComputeST(processor_count, t, fpred):
    # Get the number of tasks
    tasks = len(adj_mat)
    # Initialize a list to track the readiness of each task
    ready_task = [False] * tasks
    # The first task is always ready to be scheduled
    ready_task[0] = True
    # Initialize the ST (Start Time) matrix with a large value
    st = [[sys.maxsize] * tasks for _ in range(processor_count)]
    # Initialize the CT (Completion Time) matrix with zeros
    ct = [[0] * tasks for _ in range(processor_count)]
    # Compute ST and CT for each task in each processor
    for processor in range(processor_count):
        for ni in range(tasks):
            # Skip tasks that are not ready for scheduling
            if not ready_task[ni]:
                continue
            if ni == 0:
                # The first task starts at time 0
                st[processor][ni] = 0
            else:
                # Find the predecessor task(s) and select the first one
                nq = fpred[ni]
                nj = nq[0]
                # Compute the maximum value between the completion time of the predecessor
                # task and the start time of the predecessor task on the current processor
                max_val = max([ct[processor][nj], st[processor][nj]]) if nj is not None else 0
                st[processor][ni] = max_val
            # Compute the completion time of the current task on the current processor
            ct[processor][ni] = st[processor][ni] + t[ni][processor]
            # Update the readiness of successor tasks
            for successor in range(tasks):
                if adj_mat[ni][successor] > 0:
                    # If there is a successor, mark it as ready for scheduling
                    ready_task[successor] = True
            # Mark the current task as finished (not ready anymore)
            ready_task[ni] = False
    return st, ct
```

```python
def HTschedule(adj_mat, processors , est , ect , fpred):
    tasks = len(adj_mat)
    ready = [False] * tasks
    ready[0] = True
    rdy = [0] * tasks
    ready[0] = True
    for na in range(tasks):
        if not ready[na]:
            continue
        # Step 3: Compute fpred, CP RED, fpred0, est, ect, rdy of na
        cp_red = [0] * tasks
        fpred0 = []
        for i in range(tasks):
            if adj_mat[i][na] > 0:
                fpred[na].append(i)
                fpred0.append(i)
                cp_red[i] = adj_mat[i][na]   
        ect[na] = est[na] + cp_red[na]
        rdy[na] = ect[na]
        # Step 4: Compute cluster C(na)
        cluster = [na] + fpred[na]
        cluster = list(set(cluster))
        # Step 5: Check SUCC(na) for ready tasks
        for i in range(tasks):
            if adj_mat[na][i] > 0:
                ready[i] = True
        # Step 6: Set na 'finished'
        ready[na] = False
    # Step 7: Erase included clusters
    included_clusters = set()
    for na in range(tasks):
        if not ready[na]:
            included_clusters.update(set(fpred[na]))
    # Step 8: Map cluster C(n!) to processor P1
    processor_assignment = [0] * tasks
    for cluster_id, na in enumerate(included_clusters):
        processor_assignment[na] = 0
    p = 1
    for j in sorted(range(tasks), key=lambda x: rdy[x], reverse=True):
        # Step 9: p = p + 1
        p += 1
        # Step 10: Map cluster Cj to processor Pp
        processor_assignment[j] = p
    # Step 12: Compute ST(p)
    ComputeST(processors , processor_time , fpred)
    return processor_assignment
```
